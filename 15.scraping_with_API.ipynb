```python
def get_news_list():
    results_list = []
    results = driver.find_elements_by_xpath(news_item_xpath)
    for item in results :
        subject = item.find_element_by_class_name('subject').text
        date = item.find_element_by_class_name('date').text
        abs_text = item.find_element_by_class_name('txt').text
        print(subject, date, abs_text, sep='\n')
        results_list.append((
            subject,
            date,
            abs_text,
        ))
    return results_list
```

```python
driver.execute_script("fnPagingMove(2);")
```

```python
# if elements doesn't load correctly, wait 5seconds
driver.implicitly_wait(5)

all_results = []
for i in range(1,10+1):
    driver.execute_script("fnPagingMove({page_num});".format(page_num = i))
    item_data = get_news_list()
    all_results.append(tuple(item_data))
    sleep(1)
```
