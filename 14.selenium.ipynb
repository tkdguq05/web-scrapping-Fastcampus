## Scrapping Stackoverflow 'python' search results

```python
term = 'python'
domain = 'https://stackoverflow.com'
url = domain+'/questions/tagged/'+ term
```

```python
response = requests.get(url)
```

```python
html = response.text
```

```python
html
```

```python
soup = BeautifulSoup(html, 'lxml')
```

```python
question_section = soup.find('div',attrs={'id':'questions'})
```

```python
question_section
```

```python
questions = question_section.find_all('div', attrs={'class':'summary'})
```

```python
title_link =questions[0].find('a',attrs={'class':'question-hyperlink'})
```

```python
title = title_link.text
link = domain + title_link['href']
```

```python
print(title, link)
```

```python
abstract_text = questions[0].find('div',attrs = {'class':'excerpt'}).text
abstract_text.strip(' \r\n')
print(abstract_text)
```

```python
for question in questions :
    title_link =question.find('a',attrs={'class':'question-hyperlink'})
    title = title_link.text
    link = domain + title_link['href']
    abstract_text = question.find('div',attrs = {'class':'excerpt'}).text
    abstract_text.strip(' \r\n')
    print(abstract_text)
```
