#### 조건에 맞는 값을 가져와보자(보증금이 2000이하 거나 월세가 50이하인 매물 검색)

#### 조건 2개 사용하기 $or

```python
query = {
    '$or':[
        {
    'item.deposit':{
    '$lte':2000,
    '$gte':1000,
    }
        },
    {'item.rent':{
        '$lte':50,
        '$gte':20
    }
    }
    ]
}
print(len(list(bigbang_collection.find(query))))
list(bigbang_collection.find(query))
```
